```python
# Convert 'Size' column to numeric values
# Replace 'M' with 'e+6' for MB to Bytes and 'K' with 'e+3' for KB to Bytes
df1["Size"] = pd.to_numeric(
    df1['Size'].str.replace('M', 'e+6').str.replace('k', 'e+3'),
    errors='coerce'
)

# Convert Bytes to KB
df1["Size"] = df1["Size"] / 1000

# Fill NaN values with 'Varies with device'
df1["Size"].fillna("Varies with device", inplace=True)

# Define a function to categorize sizes into 10 categories
def categorize_size(size):
    if size == "Varies with device":
        return "Varies with device"
    elif size < 100:  # less than 0.1MB
        return "Less than 0.1MB"
    elif 100 <= size < 1000:  # 0.5MB to 1MB
        return "0.1MB to 1MB"
    elif 1000 <= size < 5000:  # 1MB to 5MB
        return "1MB to 5MB"
    elif 5000 <= size < 10000:  # 5MB to 10MB
        return "5MB to 10MB"
    elif 10000 <= size < 20000:  # 10MB to 20MB
        return "10MB to 20MB"
    elif 20000 <= size < 50000:  # 20MB to 50MB
        return "20MB to 50MB"
    elif 50000 <= size < 100000:  # 50MB to 100MB
        return "50MB to 100MB"
    else:  # greater than 100MB
        return "Greater than 100MB"

# Apply the function to create a new column 'Size Category'
df1['Size Category'] = df1['Size'].apply(categorize_size)


# Optionally, check the distribution of the size categories
df1['Size Category'].value_counts()

```

```python
# Identify columns with 'object' data type
object_columns = df1.select_dtypes(include=['object']).columns

# Loop through each column with 'object' data type
for col in object_columns:
    # Convert text to lowercase and strip leading/trailing whitespace
    df1[col] = df1[col].apply(lambda x: x.lower().strip() if isinstance(x, str) else x)
```

#### **Using Z-Score**
- The Z-score can help identify outliers by measuring how far away a data point is from the mean.


```python
from scipy import stats

# Calculate Z-scores for numerical columns
z_scores = stats.zscore(app[numerical_columns].dropna())

# Identify outliers (abs(Z-score) > 3 is a common threshold)
outliers = (abs(z_scores) > 3).any(axis=1)

# Reset index of outliers to align with the original DataFrame 'app'
outliers = outliers.reindex(app.index, fill_value=False)

# Display outliers
app_outliers = app[outliers]
app_outliers
```

#### **Using IQR Method**

- The IQR method is effective for detecting outliers by considering values beyond the typical range.


```python
# Identify outliers using IQR method
def find_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    return ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))

# Apply the IQR method to numerical columns
outliers_iqr = app[numerical_columns].apply(find_outliers_iqr)

# Display rows with outliers
app_outliers_iqr = app[(outliers_iqr.any(axis=1))]
app_outliers_iqr
```

##**Reason why we should not remove outliers during exploratory data analysis (EDA):**



### 1. Rating Column

**Example Scenario**:
- Apps in the dataset have ratings ranging from 2.0 to 4.8.
- An outlier might be an app with a rating of 1.0 or 5.0.

**Reasons Not to Remove Outliers**:
- **Representative of Extreme User Feedback**: Outliers like a rating of 1.0 or 5.0 reflect extreme user experiences that are critical for understanding user sentiment towards apps.
- **Impact on App Evaluation**: Removing extreme ratings could distort the analysis of app quality metrics, such as average ratings across categories or genres.
- **Insight into User Preferences**: Extreme ratings provide insights into highly polarizing apps, helping to identify which apps resonate strongly with users and which do not.

### 2. Reviews Column

**Example Scenario**:
- App A has 1,000,000 user reviews.
- App B has 500,000 user reviews.
- App C has 10 user reviews (potential outlier).

**Reasons Not to Remove Outliers**:
- **User Engagement and Popularity**: App C with significantly fewer reviews may indicate a new or niche app that hasn't yet gained widespread attention but could be emerging in popularity.
- **Understanding User Feedback**: Analyzing why certain apps have lower review counts offers insights into user engagement patterns and app usability.
- **Market Positioning**: Retaining outliers helps in understanding the market position of apps with varying levels of user interaction, essential for competitive analysis and strategic decision-making.

### 3. Installs Column

**Example Scenario**:
- App X has 10,000,000 installs.
- App Y has 5,000,000 installs.
- App Z has 100 installs (potential outlier).

**Reasons Not to Remove Outliers**:
- **Market Segmentation**: App Z with fewer installs may target a specific niche market or demographic that is underserved by more popular apps.
- **Emerging Trends**: Identifying apps with lower install counts can reveal emerging trends or new entrants in the app market that have potential for growth.
- **Strategic Insights**: Retaining outliers provides insights into market dynamics and consumer behavior, supporting innovation and market expansion strategies.

### 4. Price Column

**Example Scenario**:
- Most apps are free (Price = 0).
- Some apps are paid with prices ranging from $0.99 to $9.99.
- An outlier might be an app priced at $99.99.

**Reasons Not to Remove Outliers**:
- **Revenue and Monetization**: High-priced apps may cater to niche markets willing to pay premium prices for specialized features or content, contributing significantly to revenue generation.
- **Understanding Pricing Strategies**: Analyzing outlier prices offers insights into app monetization models and competitive pricing strategies.
- **Market Dynamics**: Retaining outliers ensures a comprehensive view of the competitive landscape, especially in markets where pricing diversity is a strategic advantage.

```python
# Filter top genres and subset the data
genre_counts = app['Genres'].value_counts().sort_values(ascending=False)
top_genres = genre_counts.head(10)
top = app[app['Genres'].isin(top_genres.index)]

# Set up the figure and axes with larger size
plt.figure(figsize=(18, 14))
plt.subplots_adjust(hspace=0.5)

# Bar plot for Number of Apps per Genre
plt.subplot(2, 2, 1)
sns.barplot(x=top_genres.index, y=top_genres.values, palette='viridis')
plt.title('Number of Apps per Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Apps')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Bar plot for Average Ratings across genres
plt.subplot(2, 2, 2)
sns.barplot(x='Genres', y='Rating', data=top, palette='Blues', ci=None)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.title('Average Ratings across Genres', fontsize=14)
plt.xlabel('Genres', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)

# Annotate each bar with the value
for index, value in enumerate(top.groupby('Genres')['Rating'].mean().round(2)):
    plt.text(index, value + 0.1, str(value), ha='center', va='bottom', fontsize=9)

# Bar plot for Average Reviews across genres
plt.subplot(2, 2, 3)
sns.barplot(x='Genres', y='Reviews', data=top, palette='Oranges', ci=None)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.title('Average Reviews across Genres', fontsize=14)
plt.xlabel('Genres', fontsize=12)
plt.ylabel('Average Reviews', fontsize=12)

# Annotate each bar with the value
for index, value in enumerate(top.groupby('Genres')['Reviews'].mean().round()):
    plt.text(index, value + 50000, str(value), ha='center', va='bottom', fontsize=9)

# Bar plot for Average Installs across genres
plt.subplot(2, 2, 4)
sns.barplot(x='Genres', y='Installs', data=top, palette='Greens', ci=None)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.title('Average Installs across Genres', fontsize=14)
plt.xlabel('Genres', fontsize=12)
plt.ylabel('Average Installs', fontsize=12)

# Annotate each bar with the value
for index, value in enumerate(top.groupby('Genres')['Installs'].mean().round()):
    plt.text(index, value + 1000000, str(value), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
